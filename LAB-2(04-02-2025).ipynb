```python
#12. Write a function to find min and max values of any column


import pandas as pd

def find_min_max(df, column_name):
    """
    Finds the minimum and maximum values of a specified column in a DataFrame.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The column to find min/max values.

    Returns:
        tuple: (min_value, max_value)
    """
    if column_name in df.columns:
        min_value = df[column_name].min()
        max_value = df[column_name].max()
        return min_value, max_value
    else:
        return f"Column '{column_name}' not found in DataFrame."

# Load the CSV file into a DataFrame
df = pd.read_csv('Car_sales.csv')

# Example: Finding min and max of the 'Price' column
min_price, max_price = find_min_max(df, 'Fuel capacity')

print(f"Minimum fuel capacity: {min_price}")
print(f"Maximum fuel capacity: {max_price}")
```

```python
#import pandas as pd

def find_min_max(df, column_name):
    """
    Finds the minimum and maximum values of a specified column in a DataFrame,
    handling missing values represented as '.'.
    
    Parameters:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The column to find min/max values.
    
    Returns:
        tuple: (min_value, max_value) or an error message.
    """
    if column_name in df.columns:
        # Replace '.' with NaN
        df[column_name] = df[column_name].replace(".", pd.NA)
        
        # Convert column to numeric if possible
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
        
        # Find min and max, ignoring NaN values
        min_value = df[column_name].min(skipna=True)
        max_value = df[column_name].max(skipna=True)

        return min_value, max_value
    else:
        return f"Column '{column_name}' not found in DataFrame."


# Example: Finding min and max of the 'Price' column
min_price, max_price = find_min_max(df, 'Engine size')

print(f"Minimum Price: {min_price}")
print(f"Maximum Price: {max_price}")
```
